# Manually iterations for gradient descent where slope/m is constant

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
# sample=number of rows
# features = input column
# target=output column
# informative=This parameter controls the number of informative features in the dataset. Informative features are those that actually influence the target variable.
```

# Now our target is to reach this line using gradient descent 

```python
# lets apply gradient descent assuming slope is constant m=78.35
# lets assume starting value of b is 0
```

```python
y_pred=((78.35*x)+0).reshape(4)
# y=(m*x)+b
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="Red",label="OLS")
plt.plot(x,y_pred,color="black",label="b=0")
plt.legend()
plt.show()
```

```python
# 1st iteration

m=78.35
b=0

# now we will update the value of b by calculation the value of loss_slope

loss_slope  =  -2 * np.sum(y - ( m*x.ravel() ) -b )
loss_slope
```

```python
# lets assume learnig rate =0.1

learning_rate=0.1

step_size=loss_slope*learning_rate
step_size
```

```python
# calculating new intercept

b=b-step_size
b
```

```python
y_pred1=((m*x)+b).reshape(4) # y=(m*x)+b
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="Red",label="OLS")
plt.plot(x,y_pred,color="black",label="b=0")
plt.plot(x,y_pred1,color='green',label="b=20.927763408209216")
plt.legend()
plt.show()

# as shown through 1 iteration our best fit line moves towards the line that is created using ols
```

```python
# 2nd iteraton
```

```python
loss_slope  =  -2 * np.sum(y - ( m*x.ravel() ) -b )
loss_slope
```

```python
learning_rate=0.1

step_size=loss_slope*learning_rate
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((m*x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="Red",label="OLS")
plt.plot(x,y_pred,color="black",label="b=0")
plt.plot(x,y_pred1,color='green',label="b=20.927763408209216")
plt.plot(x,y_pred2,color='orange',label="b=25.11331608985106")
plt.legend()
plt.show()
# our line moved more closer to ols line
```

```python
# 3rd iteration
```

```python
loss_slope=-2 * np.sum(y-(m*x.ravel())-b)
loss_slope
```

```python
step_size=learning_rate*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((m*x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="Red",label="OLS")
plt.plot(x,y_pred,color="black",label="b=0")
plt.plot(x,y_pred1,color='green',label="b=20.927763408209216")
plt.plot(x,y_pred2,color='orange',label="b=25.11331608985106")
plt.plot(x,y_pred3,color='pink',label="b=25.95042662617943")
plt.legend()
plt.show()

# moved almost on ols line
```

```python
# we started with be as -100 and still we reached near best fit line

b=-100
m=78.35
lr=0.1
epochs=10

for i in range (epochs):
    
    loss_slope= -2 * np.sum(y-(m*x.ravel())-b)
    
    b=b-(lr*loss_slope)
    
    y_pred=(m*x)+b
    
    plt.plot(x,y_pred)

    
plt.scatter(x,y)
```

# Making class of gradient descent where slope/m is constant

```python
x,y=make_regression(n_samples=100,n_features=1,n_targets=1,n_informative=1,noise=20,random_state=24)
```

```python
plt.scatter(x,y)
```

```python
# finding slope and intercept using ols
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
lr.coef_    # slope
```

```python
lr.intercept_ # intercept
```

```python
class gdregressor:
    
    def __init__(self,learning_rate,epochs):
        self.m=73.40  # constant value
        self.b=-120   # random value
        self.lr=learning_rate
        self.epochs=epochs
    
    def fit(self,x,y):
        
        # calculate value of b using gradient decent
        for i in range (self.epochs):
            loss_slope= -2 * np.sum (y - self.m * x.ravel() - self.b)
            self.b=self.b - (self.lr*loss_slope)
            print('loss_slope is:',loss_slope,'and intercept is:',self.b)
        print("The final value of intercept is:",self.b)
```

```python
gd=gdregressor(0.001,40)
```

```python
gd.fit(x,y)
```
