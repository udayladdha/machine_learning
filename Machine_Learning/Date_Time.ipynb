# Working with dates

```python
# converting dtype into datetime
date["date"]=pd.to_datetime(date["date"])
```

```python
date.info()
```

# 1. extracting year

```python
date["date_year"]=date["date"].dt.year
date.head()
```

# 2.extract month

```python
date["date_month"]=date["date"].dt.month
date.head()
```

# 3. extract day

```python
date["date_day"]=date["date"].dt.day
date.head()
```

# 4. extract month name

```python
date["date_month_name"]=date["date"].dt.month_name()
date.sample(5)
```

# 5. extracting day name

```python
date["date_day_name"]=date["date"].dt.day_name()
date.sample(5)
```

# 5. extracting the day of week 

```python
date["date_day_of_week"]=date["date"].dt.dayofweek
date.sample(5)    # monday starts from 0
```

# 6. is weekend or not

```python
date["date_is_weekend"]=np.where(date["date_day_name"].isin(["Saturday","Sunday"]),1,0)
# date.drop(columns=["product_id","city_id","orders"],inplace=True)
date.sample(5)
```

# 7.Exract week of the year

```python
date["date_week_of_the_yeay"]=date["date"].dt.weekofyear
date.sample(5)  # total 52 weeks are there in a year
```

# 8.Extracting quater

```python
date["date_quater"]=date["date"].dt.quarter
date.sample(5)
```

# 9.extracting semester

```python
date["date_semester"]=np.where(date["date_quater"].isin([1,2]),1,2)
date.sample(5)  # 6 months is 1 semester
```
