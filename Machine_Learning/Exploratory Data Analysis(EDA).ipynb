# 1.Scatterplot (Numerical-Numerical)

```python
# scatter plot works on numerical columns
# it tells the relationship between 2 numerical columns
# it has hue,style,size parameter which allow us to do multivariate analysis as shown below
```

```python
# multivariate analysis using scatter plot
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="smoker",size="size")
```

# 2. Bar plot (Categorical-Numerical)

```python
# Bar plot is used to represent categorical and numerical data
# generally we use categorical dat aon x axis and numerical data on y axis
# this black line on the bar shows confidence interval
```

```python
# multivariate analysis using bar plot
sns.barplot(data=titanic,x="Pclass",y="Age",hue="Sex",)
```

# 3.Box Plot (Categorical-Numerical)

```python
sns.boxplot(data=titanic,x="Sex",y="Age",hue="Survived")
```

# 4.Dist plot(Numerical - categorical)

```python
# Dist plot does not have hue parameter so we created a method through which we can do multvariate analysis using dist plot
# below method is telling us the probablity of people's age who survived and who does not survived 
```

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"],hist=False,color="black")
sns.distplot(titanic[titanic["Survived"]==1]["Age"],hist=False,color="orange")
```

# 5.Heat map(Categorical - categorical)

```python
titanic.head()
```

```python
# Cross tab is a contingency table that shows the relation between 2 or more categorical values 
# It tells us how the data is distributed and related to each other among the given categorical columns
# It has a margin parameter which gives us the total of the rows
pd.crosstab(titanic["Pclass"],titanic["Survived"],margins=True)
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
# Heat maps only provide us the number , to find it in percentage we con do the following steps
```

```python
# percentage of people in pcalss that survived
((titanic.groupby("Pclass")["Survived"].mean())*100)
```

```python
((titanic.groupby("Pclass")["Survived"].mean())*100).plot(kind="bar")
```

# 6.Cluster map (categorical - categorical)

```python
# cluster map is used to tell the clustering structure within given data sets
```

```python
pd.crosstab(titanic["Parch"],titanic["Survived"],margins=True)
```

```python
sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
```
