    1.mse
    2.rmse
    3.mae

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("//Users//udayladdha//Desktop//Data science/DataSets//placement.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=df["cgpa"],y=df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[["cgpa"]]
y=df[["package"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,accuracy_score,mean_absolute_error,mean_squared_error
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("mse:",mean_squared_error(y_pred,y_test)) # mean squared error 
# unit is squared as output column 
# for eg out y is lpa so  so it tells us error square root of 0.08 lpa
```

```python
# this is the best fit line that is created by our model 
plt.scatter(x=df["cgpa"],y=df["package"])
plt.xlabel("cgpa")
plt.ylabel("package")
plt.plot(x_train,lr.predict(x_train),color="red")
```

```python
print("mae:",mean_absolute_error(y_pred,y_test)) # mean absolute error
# unit is same as output column 
# for eg out y is lpa so  so it tells us error of 0.08 lpa 
```

```python
print("r2score:",r2_score(y_pred,y_test))  # it tells us how much our regression line is better than. mean line
```

```python
r2=r2_score(y_pred,y_test)
```

```python
print("rmse:",np.sqrt(mean_absolute_error(y_pred,y_test))) # root of mean squared error
```
